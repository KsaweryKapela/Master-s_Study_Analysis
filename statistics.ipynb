```python
import pandas as pd

file_name = 'processing_ready_data.xlsx'
df = pd.read_excel(io=file_name)
```

```python
pd.options.display.max_columns = 50
df.drop(labels=['Unnamed: 0'], axis=1)
df[:5]
```

```python
def negative_bold(val):
    bold = 'bold' if 1 > abs(val) > 0.2 else ''
    return f'font-weight: {bold}'

correlation_items = ['extreme_score', 'PPCS_score', 'weekly_minutes', 'weekly_session', 'session_time', 'well-being',
                     'father_relations', 'mother_relations', 'romantic_life', 'sexual_life', 'reenacting', 'age']
```

```python
correlations_df = df[correlation_items].corr(method='spearman')

correlations_df = correlations_df[:5].style.applymap(negative_bold)

correlations_df
```
