```python
import pandas as pd
import numpy as np

file_name = 'processing_ready_data.xlsx'
df = pd.read_excel(io=file_name)
```

```python
extreme_columns = df.columns[2:10]
```

```python
x_df = df[extreme_columns]
```

```python
x_df = x_df.apply(lambda x: x - 1)
```

```python
sum = x_df.iloc[:].sum(axis=1)
x_df.insert(len(x_df.columns), 'Sum', sum)
x_df
```

```python
extreme_proportion = [f'{item}_pro' for item in extreme_columns]
extreme_proportion
```

```python
extreme_proportion[0][:-4]
```

```python
for item in extreme_proportion:
    x_df[item] = x_df[item[:-4]].div(x_df['Sum'].values)
```

```python
values = [1.21, 1.49, 1.75, 2.06, 2.53, 2.85, 3.19, 3.78]
extreme_result = [f'{item}_res' for item in extreme_columns]
```

```python
for i in range(len(values)):
    print(extreme_result[i])
    x_df[extreme_result[i]] = x_df[extreme_proportion[i]].mul(values[i])
```

```python
score = x_df[extreme_result].iloc[:].sum(axis=1)
x_df.insert(0, 'Score', score)
```

```python
for x, y in zip(df['extreme_score'], x_df['Score']):
    print(x, y)
```
