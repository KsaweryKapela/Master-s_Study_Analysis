```python
import pandas as pd

file_name = 'processing_ready_data.xlsx'
df = pd.read_excel(io=file_name)

df_no_outliers = df[df.weekly_minutes < 500]
```

```python
pd.options.display.max_columns = 50
df.drop(labels=['Unnamed: 0'], axis=1)
df[:5]
```

```python
def negative_bold(val):
    bold = 'bold' if 1 > abs(val) > 0.2 else ''
    return f'font-weight: {bold}'

correlation_items = ['sexual_life', 'extreme_score', 'PPCS_score', 'weekly_minutes', 'weekly_session', 'session_time',
                     'reenacting', 'well-being', 'friends_relations', 'romantic_life', 'age', 'gender']
```

```python
def get_correlations(condition, method='pearson'):
    correlations_df = condition
    correlations_df = correlations_df[correlation_items].corr(method=method)
    correlations_df = correlations_df[:4].style.applymap(negative_bold)
    return correlations_df
```

```python
#EVERYONE

correlations_df = get_correlations(df)
correlations_df
```

```python
#EVERYONE WITHOUT OUTLIERS

correlations_df = get_correlations(df_no_outliers)
correlations_df
```

```python
# MAN ONLY

correlations_male = get_correlations(df.loc[df['gender'] == 'Mężczyzna'])
correlations_male
```

```python
# WOMAN ONLY

correlations_female = get_correlations(df.loc[df['gender'] == 'Kobieta'])
correlations_female
```

```python
# OLDER THAN 23

correlations_older = get_correlations(df.loc[df['age'] > 23])
correlations_older
```

```python
# YOUNGER THAN 23

correlations_younger = get_correlations(df.loc[df['age'] < 23])

correlations_younger
```

```python
# YOUNGER THAN 23 WITHOUT OUTLIERS

correlations_younger = get_correlations(df_no_outliers.loc[df_no_outliers['age'] < 23])


correlations_younger
```

```python
# SEXUAL LIFE AND WEEKLY MINUTES CORRELATION IN SUB 23 AGE SAMPLE WITHOUT OUTLIERS
import scipy.stats

correlation = scipy.stats.pearsonr(df_no_outliers.loc[df_no_outliers['age'] < 23].sexual_life,
                                   df_no_outliers.loc[df_no_outliers['age'] < 23].weekly_minutes)

x = [round(corr, 4) for corr in correlation]
print(f'R={x[0]}, P={x[1]}')
```
