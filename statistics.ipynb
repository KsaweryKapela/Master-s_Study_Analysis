```python
import pandas as pd

file_name = 'processing_ready_data.xlsx'
df = pd.read_excel(io=file_name)
```

```python
pd.options.display.max_columns = 50
df.drop(labels=['Unnamed: 0'], axis=1)
df[:5]
```

```python
def negative_bold(val):
    bold = 'bold' if 1 > abs(val) > 0.2 else ''
    return f'font-weight: {bold}'

correlation_items = ['extreme_score', 'PPCS_score', 'sexual_life', 'weekly_session', 'session_time', 'reenacting', 'well-being',
                     'father_relations', 'mother_relations', 'romantic_life', 'age', 'gender']
```

```python
#EVERYONE

correlations_df = df[correlation_items].corr(method='spearman')

correlations_df = correlations_df[:6].style.applymap(negative_bold)

correlations_df
```

```python
# MAN ONLY

correlations_male = df.loc[df['gender'] == 'Mężczyzna']
correlations_male = correlations_male[correlation_items].corr(method='pearson')

correlations_male = correlations_male[:6].style.applymap(negative_bold)
correlations_male
```

```python
# WOMAN ONLY

correlations_female = df.loc[df['gender'] == 'Kobieta']
correlations_female = correlations_female[correlation_items].corr(method='pearson')

correlations_female = correlations_female[:6].style.applymap(negative_bold)
correlations_female
```

```python
# OLDER THAN 23

correlations_age = df.loc[df['age'] > 23]
correlations_age = correlations_age[correlation_items].corr(method='pearson')

correlations_age = correlations_age[:6].style.applymap(negative_bold)
correlations_age
```

```python
# YOUNGER THAN 23

correlations_age = df.loc[df['age'] < 23]
correlations_age = correlations_age[correlation_items].corr(method='pearson')

correlations_age = correlations_age[:6].style.applymap(negative_bold)
correlations_age
```

```python
# BELIEVE IN GOD

correlations_age = df.loc[df['believes_in_god'] == True]
correlations_age = correlations_age[correlation_items].corr(method='spearman')

correlations_age = correlations_age[:6].style.applymap(negative_bold)
correlations_age
```

```python
# DOESN'T BELIEVE IN GOD

correlations_god = df.loc[df['believes_in_god'] == False]
correlations_god = correlations_god[correlation_items].corr(method='spearman')

correlations_god = correlations_god[:6].style.applymap(negative_bold)
correlations_god
```

```python
# IN RELATIONSHIP

correlations_age = df.loc[df['in_relationship'] == True]
correlations_age = correlations_age[correlation_items].corr(method='pearson')

correlations_age = correlations_age[:6].style.applymap(negative_bold)
correlations_age
```

```python
# NOT IN RELATIONSHIP

correlations_age = df.loc[df['in_relationship'] == False]
correlations_age = correlations_age[correlation_items].corr(method='pearson')

correlations_age = correlations_age[:6].style.applymap(negative_bold)
correlations_age
```
