```python
import pandas as pd

file_name = 'processing_ready_data.xlsx'
df = pd.read_excel(io=file_name)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

df.PPCS_score.plot(kind='hist', color='purple', edgecolor='black', figsize=(10,7))
plt.title('Distribution of PPCS scores', size=24)
plt.xlabel('PPCS score', size=18)
plt.ylabel('Frequency', size=18)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

df.extreme_score.plot(kind='hist', color='purple', edgecolor='black', figsize=(10,7))
plt.title('Distribution of extreme scores', size=24)
plt.xlabel('Extreme score', size=18)
plt.ylabel('Frequency', size=18)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

df.sexual_life.plot(kind='hist', color='purple', edgecolor='black', figsize=(10,7))
plt.title('Distribution of sexual life satisfaction', size=24)
plt.xlabel('Sexual life satisfaction', size=18)
plt.ylabel('Frequency', size=18)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

df.reenacting.plot(kind='hist', color='purple', edgecolor='black', figsize=(10,7))
plt.title('Distribution of wanting to reenact scenes from pornography', size=24)
plt.xlabel('Reenacting', size=18)

plt.ylabel('Frequency', size=18)
```

```python
from matplotlib.pyplot import figure

sorted_minutes = sorted(df.weekly_minutes)
figure(figsize=(10, 7), dpi=80)
plt.plot(sorted_minutes, 'o')
plt.title('Weekly minutes of pornography watching', size=24)
plt.xlabel('Respondents')
plt.ylabel('Weekly minutes')
```

```python
from matplotlib.pyplot import figure
df_cleared = df[df.weekly_minutes < 500]

sorted_minutes = sorted(df_cleared.weekly_minutes)

figure(figsize=(10, 7), dpi=80)
plt.plot(sorted_minutes, 'o')
plt.title('Weekly minutes of pornography watching (without egde-cases)', size=20)
plt.xlabel('Respondents')
plt.ylabel('Weekly minutes')
```
